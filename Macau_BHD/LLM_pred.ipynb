```python
import openai
import os
import time

os.environ.get("OpenAI API key") 

def LLM_pred(prior,few_shot_examples,user_description):
    
    prompt_content = "You are a geological expert. Your task is to label the given descriptions with layer type based on the provided examples."

    RAGtask_content =RAGtask_content = f"""


    Use the following examples to assist with your task:
    {few_shot_examples}

    Label for the given description in the format `<description>,<layer>`:
    {user_description}
    """

    messages = [
        {"role": "system",
        }
    ]

    prompt_messages = [
        {
            "role": "system", 
            "content": prompt_content  
        }
    ]

    user_message = {
            "role": "user",
            "content": RAGtask_content
        }

    prompt_messages.append(user_message)

    max_retries = 3
    retries = 0
    success = False

    while retries < max_retries and not success:
        try:
            response = openai.ChatCompletion.create(
                messages=prompt_messages,
                temperature=0,
                #max_tokens=9999,
                model="gpt-4o",
                
            )
            success = True
        except openai.error.APIConnectionError as e:
            print(f"Connection error: {e}")
            retries += 1
            time.sleep(2)  
    if success:
        print(f"API call succeeded. Returned {len(response.choices)} choices")
        
        for idx, choice in enumerate(response.choices, 1):
            print(f"\n===== Choice {idx} =====\n{choice.message.content}")
    else:
        print("API call failed after retries.")

    return choice.message.content
```
